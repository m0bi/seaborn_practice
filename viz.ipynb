```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("~/seaborn_practice/FAO.csv", encoding="ISO-8859-1")
```

```python
df.head()
```

```python
top_df = df.nlargest(20, "Y2013")
top_df.head(20)
```

```python
plot_one = sns.factorplot(x="Area", data=top_df, kind="count", palette="BuPu", size=6, aspect=1.5)
```

```python
df_first_specific = df[df.Item=='Coffee and products'].copy()
df_first_specific.loc[:, "Percentage"] = df_first_specific.Y2013/df_first_specific.Y2013.sum()
df_first_specific = df_first_specific.sort_values(by="Y2013", ascending=False)
df_first_specific.loc[:, "Cumsum"] = df_first_specific.Percentage.cumsum()
```

```python
top_coffee = df_first_specific.nlargest(5, "Y2013")
top_coffee.head()
```

```python
sns.set({"figure.figsize":(12,12)})
plt.xticks(rotation=70)
sns.barplot(x="Area", y="Y2013", data=top_coffee, label="item")
```

```python
top_coffee_2012 = df_first_specific.nlargest(5, "Y2012")
top_coffee_2012.head()
```

```python
sns.set(font_scale=1.5)
sns.set_color_codes("pastel")
sns.barplot(x="Area", y="Y2013", data=top_coffee, label="2013", color="r")
sns.barplot(x="Area", y="Y2012", data=top_coffee_2012, label="2012", color="b")

ax = plt.gca()
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlim=(0,6), ylabel="Coffee", xlabel="Country")
sns.despine(left=True, bottom=True)
```

```python
dist_df = df.nlargest(1000, "Y2013")

sns.set(rc={'figure.figsize': (18,18)})
sns.swarmplot(x="Item", y="Y2013", data=dist_df)
plt.xticks(rotation=90)
```
